```python
def factorial(n):
    if n==1:
      return 1
    return factorial(n-1)*n
print(factorial(5))
```

```python
#List Comprehension Speed‑Up
import timeit
import matplotlib.pyplot as plt

def loop_append():
    squares1 = []
    for x in range(1, 100001):
        squares1.append(x**2)

def list_comprehension():
    squares2 = [x**2 for x in range(1, 100001)]

time_loop = timeit.timeit(loop_append, number=10)
time_comp = timeit.timeit(list_comprehension, number=10)

print(f"For loop with append: {time_loop:.4f} seconds")
print(f"List comprehension:   {time_comp:.4f} seconds")

plt.bar(["For Loop", "List Comprehension"], [time_loop, time_comp])
plt.ylabel("Time (seconds)")
plt.title("For Loop vs List Comprehension Runtime")
plt.show()
```
